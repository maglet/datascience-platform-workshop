### Loading the data

### Scatterplot

To make a scatterplot, use `replot` and specify what dataset you're using (like with `distplot`) and what variables you want to plot on the `x` and `y` axes.

```python
sns.relplot(data=surveys,
            x="weight", 
            y="hindfoot_length", 
            alpha = 0.1, 
            hue = "species_id")
```
